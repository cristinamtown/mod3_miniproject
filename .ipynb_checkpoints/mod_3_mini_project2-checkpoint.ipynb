#### Columns names and descriptions

There are 25 variables:

**ID**: ID of each client
   
**LIMIT_BAL**: Amount of given credit in NT dollars (includes individual and family/supplementary credit

**SEX**: Gender (1=male, 2=female)
    
**EDUCATION**: (1=graduate school, 2=university, 3=high school, 4=others, 5=unknown, 6=unknown)

**MARRIAGE**: Marital status (1=married, 2=single, 3=others)

**AGE**: Age in years

**PAY_0**: Repayment status in September, 2005 (-1=pay duly, 1=payment delay for one month, 2=payment delay for two months, â€¦ 8=payment delay for eight months, 9=payment delay for nine months and above)

**PAY_2**: Repayment status in August, 2005 (scale same as above)

**PAY_3**: Repayment status in July, 2005 (scale same as above)

**PAY_4**: Repayment status in June, 2005 (scale same as above)

**PAY_5**: Repayment status in May, 2005 (scale same as above)

**PAY_6**: Repayment status in April, 2005 (scale same as above)

**BILL_AMT1**: Amount of bill statement in September, 2005 (NT dollar)

**BILL_AMT2**: Amount of bill statement in August, 2005 (NT dollar)

**BILL_AMT3**: Amount of bill statement in July, 2005 (NT dollar)

**BILL_AMT4**: Amount of bill statement in June, 2005 (NT dollar)

**BILL_AMT5**: Amount of bill statement in May, 2005 (NT dollar)

**BILL_AMT6**: Amount of bill statement in April, 2005 (NT dollar)

**PAY_AMT1**: Amount of previous payment in September, 2005 (NT dollar)

**PAY_AMT2**: Amount of previous payment in August, 2005 (NT dollar)

**PAY_AMT3**: Amount of previous payment in July, 2005 (NT dollar)

**PAY_AMT4**: Amount of previous payment in June, 2005 (NT dollar)

**PAY_AMT5**: Amount of previous payment in May, 2005 (NT dollar)

**PAY_AMT6**: Amount of previous payment in April, 2005 (NT dollar)
    default.payment.next.month: Default payment (1=yes, 0=no)



#### Scrub dataset

```python
# Drop ID column
df.drop(['ID'], axis=1, inplace=True)
```

```python
import missingno
missingno.matrix(df)
df.isna().sum()
```

```python
# Convert Columns that represent binary or qauliatative variables into obj
col_str = ['SEX', 'EDUCATION', 'MARRIAGE', 'PAY_0', 'PAY_2', 'PAY_3', 'PAY_4',
           'PAY_5', 'PAY_6']
           
df[col_str] = df[col_str].astype(str)

# Quick check
df.info()
```

```python
# Defining Dictionary Map for SEX categories
SEX_map = {'1':'Male', '2':'Female'}
df['SEX'] = df['SEX'].map(SEX_map)
print(f"SEX: {df['SEX'].value_counts(dropna=False)}")

# Defining Dictionary Map for EDUCATION categories      
EDUCATION_map = {'1':'graduate school', '2':'university', 
                '3':'high school', '4':'others', '5':'unknown', '6':'unknown'}
df['EDUCATION'] = df['EDUCATION'].map(EDUCATION_map)
print(f"EDUCATION: {df['EDUCATION'].value_counts(dropna=False)}")

# Defining Dictionary Map for MARRIAGE categories      
MARRIAGE_map = {'1':'married', '2':'single', '3':'other'}
df['MARRIAGE'] = df['MARRIAGE'].map(MARRIAGE_map)
print(f"MARRIAGE: {df['MARRIAGE'].value_counts(dropna=False)}")
```

```python
# See if any nulls appeared after replacing
# missingno.matrix(df)
df.isna().sum()
```

```python
# Drop rows with null values
df.dropna(inplace=True)
df.isna().sum()
```

```python
# One hot encode object columns
feats = col_str

df = pd.get_dummies(df, drop_first=True, columns=feats)
df.info()
```

```python
# briefly explore the target
df['default.payment.next.month'].value_counts() 
```

## Functions 

```python
# # Train Test Split function
from sklearn.model_selection import train_test_split

def tts(X, y, model):
    """Performs train test split with test_size=25 and random_state=42"""
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
    return X_train, X_test, y_train, y_test
```

```python
# Function to print scores
from sklearn.metrics import precision_score, recall_score, accuracy_score, f1_score, confusion_matrix
def print_metrics(labels, preds):
    """input (labels, preds) and prints Precision score, Recall score,
    Accuracy score, F1 score, and Confusion matrix"""
    print("Precision Score: {}".format(precision_score(labels, preds)))
    print("Recall Score: {}".format(recall_score(labels, preds)))
    print("Accuracy Score: {}".format(accuracy_score(labels, preds)))
    print("F1 Score: {}".format(f1_score(labels, preds)))
    print("Confision Matrix: \n{}".format(confusion_matrix(labels, preds)))
```

### KNN

```python
# copy df as to not change anything for future models
df_knn = df.copy()

# set target variable
target = df_knn['default.payment.next.month'].astype(int)

# drop target from df
df_knn.drop('default.payment.next.month', axis=1, inplace=True)
df_knn.columns
```

```python
X_train, X_test, y_train, y_test = tts(df_knn, target)
```

```python
# from sklearn.model_selection import train_test_split

# X_train, X_test, y_train, y_test = train_test_split(df_knn, target, test_size=0.25, random_state=42)
```

```python
# Import StandardScaler
from sklearn.preprocessing import StandardScaler

# Instantiate StandardScaler
scaler = StandardScaler()

# Transform the training and test sets
scaled_data_train = scaler.fit_transform(X_train)
scaled_data_test = scaler.transform(X_test)

# Convert into a DataFrame
scaled_df_train = pd.DataFrame(scaled_data_train, columns=df_knn.columns)
scaled_df_train.head()
```

```python
# Import KNeighborsClassifier
from sklearn.neighbors import KNeighborsClassifier

def find_best_k(X_train, y_train, X_test, y_test, min_k=1, max_k=25):
    best_k = 0
    best_score = 0.0
    for k in range(min_k, max_k+1, 2):
        knn = KNeighborsClassifier(n_neighbors=k)
        knn.fit(X_train, y_train)
        preds = knn.predict(X_test)
        f1 = f1_score(y_test, preds, average='micro')
        if f1 > best_score:
            best_k = k
            best_score = f1
    
    print("Best Value for k: {}".format(best_k))
    print("F1-Score: {}".format(best_score))
```

```python
# Find best K

find_best_k(scaled_data_train, y_train, scaled_data_test, y_test)
```

```python
# Instantiate KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=25)

# Fit the classifier
clf.fit(scaled_data_train, y_train)

# Predict on the test set
test_preds = clf.predict(scaled_data_test)

```

#### Evaluate KNN model

```python
# Print scores
print_metrics(y_test, test_preds)
```
